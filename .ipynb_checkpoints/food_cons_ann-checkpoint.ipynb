```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suppose your cleaned DataFrame is called df_clean
# Visualize boxplot for food consumption
plt.figure(figsize=(10,6))
sns.boxplot(y=df_clean['food_cons_ann'])
plt.title('Boxplot of Food Consumption (food_cons_ann)')
plt.ylabel('Food Consumption')
plt.show()

# Optional: histogram to see distribution
plt.figure(figsize=(10,6))
sns.histplot(df_clean['food_cons_ann'], bins=50, kde=True)
plt.title('Distribution of Food Consumption (food_cons_ann)')
plt.xlabel('Food Consumption')
plt.ylabel('Count')
plt.show()
```

```python
Q1 = df_clean['food_cons_ann'].quantile(0.25)
Q3 = df_clean['food_cons_ann'].quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df_clean[~((df_clean['food_cons_ann'] < (Q1 - 1.5 * IQR)) | 
                            (df_clean['food_cons_ann'] > (Q3 + 1.5 * IQR)))]
```

```python

# Suppose your cleaned DataFrame is called df_clean
# Visualize boxplot for food consumption
plt.figure(figsize=(10,6))
sns.boxplot(y=df_clean['df_no_outliers'])
plt.title('Boxplot of Food Consumption (food_cons_ann)')
plt.ylabel('Food Consumption')
plt.show()

# Optional: histogram to see distribution
plt.figure(figsize=(10,6))
sns.histplot(df_clean['df_no_outliers'], bins=50, kde=True)
plt.title('Distribution of Food Consumption (food_cons_ann)')
plt.xlabel('Food Consumption')
plt.ylabel('Count')
plt.show()
```
