```python

# Drop identifier and confidential columns
cols_to_drop = [
    'household_id', 'ea_id', 'interview__key', 'interview__id',  # IDs
    'saq11', 'saq13', 'saq17', 'saq18', 'saq21',                 # confidential / unusable
    'saq19__Timestamp'                                           # missing values
]

df_clean = df.drop(columns=cols_to_drop)

# Optional: convert location columns to numeric
df_clean['saq19__Latitude'] = pd.to_numeric(df_clean['saq19__Latitude'], errors='coerce')
df_clean['saq19__Longitude'] = pd.to_numeric(df_clean['saq19__Longitude'], errors='coerce')
df_clean['saq19__Accuracy'] = pd.to_numeric(df_clean['saq19__Accuracy'], errors='coerce')
df_clean['saq19__Altitude'] = pd.to_numeric(df_clean['saq19__Altitude'], errors='coerce')

# Check result
print(df_clean.columns)
print(df_clean.shape)
print(df_clean.head())
```

```python


# Fill NaN with 0
location_cols = ['saq19__Latitude', 'saq19__Longitude', 'saq19__Accuracy', 'saq19__Altitude']
df_clean[location_cols] = df_clean[location_cols].fillna(0)

print(df_clean[location_cols].isnull().sum())  # Should show 0
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

numeric_cols = ['pw_w5','saq02','saq03','saq04','saq05','saq06','saq07','saq08','saq09',
                'saq19__Latitude','saq19__Longitude','saq19__Accuracy','saq19__Altitude']
X_numeric = df_clean[numeric_cols]

vif_data = pd.DataFrame()
vif_data["feature"] = X_numeric.columns
vif_data["VIF"] = [variance_inflation_factor(X_numeric.values, i) for i in range(X_numeric.shape[1])]

print(vif_data)
```
